```python
import random
from math import exp, log
```

```python
X = [(0,0), (1,0), (0,1), (1,1)]
Y = [0,0,0,1]
```

```python
class logistic_regression_model():
    def __init__(self):
        self.w = [random.random(), random.random()]
        self.b = random.random()
        
    def sigmoid(self, z):
        return 1/(1+exp(-z))
        
    def predict(self, x):
        z = self.w[0]*x[0]+self.w[1]*x[1]+self.b
        a = self.sigmoid(z)
        return a
```

```python
model = logistic_regression_model()
```

```python
def train(X, Y, model, lr=0.1):
    dw0 = 0.0
    dw1 = 0.0
    db = 0.0
    m = len(X)
    cost = 0.0
    for x,y in zip(X, Y):
        a = model.predict(x)
        if y==1:
            cost-=log(a)
        else:
            cost-=log(1-a)
        
        dw0+=(a-y)*x[0]
        dw1+=(a-y)*x[1]
        db+=(a-y)
        
    cost/=m
    model.w[0]-=lr*dw0/m
    model.w[1]-=lr*dw1/m
    model.b-=lr*db/m
    
    return cost 
```

```python
for epoch in range(10000):
    cost = train(X, Y, model, 0.1)
    if epoch%100==0:
        print(epoch, cost)
```
